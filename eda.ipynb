```python
import numpy as np,pandas as pd,matplotlib.pyplot as plt,seaborn as sns,sklearn as sk
```

```python
df=pd.read_csv('train.csv')
df
```

```python
df[['Deck', 'Num', 'Side']] = df['Cabin'].str.split('/', expand=True)
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
y=df['Transported']
y
```

```python
X=df.drop(['Transported','PassengerId','Name','Cabin'],axis=1)
X
```

```python

from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.compose import ColumnTransformer,make_column_selector
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder,OrdinalEncoder
from sklearn.impute import SimpleImputer,KNNImputer
from sklearn.ensemble import RandomForestClassifier
import xgboost
from xgboost import XGBClassifier
```

```python
ppl = Pipeline([
    ('cleaning', ColumnTransformer([
        # Step 2.1: Apply steps for numerical features
        ('num',make_pipeline(
            KNNImputer(),
            MinMaxScaler()),
            make_column_selector(dtype_include=['float64','int64'])
        ),
        # Step 2.2 Apply steps for nominal categorial features
        ('cat_nom',make_pipeline(
            SimpleImputer(strategy='most_frequent'),
            OneHotEncoder(sparse=False, handle_unknown='ignore',drop='first')),
            ['HomePlanet','CryoSleep','Destination','Deck','Side']
        ),
        (
            'cat_ord',make_pipeline(
                SimpleImputer(strategy='most_frequent'),
                OrdinalEncoder(categories=[[False,True]])
            ),['VIP']
        )
        ])
    ),
    ('rf',RandomForestClassifier(n_estimators=1000,max_depth=10))
])
```

```python
ppl.fit(X,y)
```

```python
df1=pd.read_csv('test.csv')
df1.head()
```

```python
df1[['Deck', 'Num', 'Side']] = df1['Cabin'].str.split('/', expand=True)
df1.drop(['PassengerId','Name','Cabin'],axis=1,inplace=True)
df1.head()
```

```python
res=ppl.predict(df1)
res
```

```python
df1['PassengerId'].values
```

```python
submission=pd.DataFrame(data={'PassengerId':df1['PassengerId'].values,'Transported':res})
submission
```

```python
submission=submission.replace({'Transported':{1:True,0:False}})
submission
```
